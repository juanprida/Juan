```python
def reemplazar(x):
    if x == 'Y':
        return 1
    return 0
```

```python
data['SHOOTING'] = data['SHOOTING'].apply(reemplazar)
```

```python
sns.scatterplot(
    data=data,
    x='HOUR',
    y='DAY_OF_WEEK',
    hue='SHOOTING'
)
```

```python
data.groupby(['DAY_OF_WEEK', 'HOUR'])['SHOOTING'].mean()
```

```python
pepe=data.groupby(['DAY_OF_WEEK', 'HOUR'])['SHOOTING'].mean().reset_index().pivot(columns='DAY_OF_WEEK', index='HOUR')
```

```python
fig, axs = plt.subplots(1, 1, figsize=(20, 10))
sns.heatmap(100*pepe, ax=axs, annot=True)
```

```python
data['OCCURRED_ON_DATE'] = pd.to_datetime(data['OCCURRED_ON_DATE']).dt.date
```

```python
tendency = data.groupby(['OCCURRED_ON_DATE'])['INCIDENT_NUMBER'].count().reset_index()
```

```python
fig, axs = plt.subplots(1, 1, figsize=(20, 20))
sns.lineplot(data=tendency, x='OCCURRED_ON_DATE', y='INCIDENT_NUMBER', ax=axs)
```

```python
filtro_ano= data[(data['YEAR']!=2015) & (data['YEAR']!=2018)]
```

```python
sns.countplot(filtro_ano['MONTH'])
```

```python
gara=(filtro_ano['MONTH'].value_counts()/(filtro_ano['MONTH'].value_counts().mean()))-1
gara= gara.reset_index()
gara= gara.rename(columns={'index':'Month','MONTH':'AVG'}).sort_values('Month')
```

```python
sns.barplot(data=gara, x='Month', y='AVG')
```
